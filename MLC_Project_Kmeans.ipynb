```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('inspection2023_Ver2.csv')
dataset
```

```python
dataset.columns
```

```python
data = dataset[['CAMIS', 'GRADE','num_foodDO', 'median income','prop_bachelor','pop_sqkm','num_crime_in2021','Hispanic or Latino','white', 'Black or African American', 'American Indian', 'Asian','Hawaiian', 'other', 'new_cuisine']]
data
```

```python
data["GRADE"].value_counts()
```

```python
X = data.loc[:,['num_foodDO', 'median income','prop_bachelor','pop_sqkm','num_crime_in2021','Hispanic or Latino','white', 'Black or African American', 'American Indian', 'Asian','Hawaiian', 'other','new_cuisine']]
Y_true = data["GRADE"]
X
```

```python
# Encode categorical variables using one-hot encoding
X = pd.get_dummies(X, columns=['new_cuisine'])
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X = ss.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
y_pred = KMeans(n_clusters=3, random_state = 234).fit_predict(X)
```

```python
for i in range(9259):
    y_pred[i]
```

```python
Y_true_list = Y_true.tolist()
```

```python
grade = data["GRADE"]
grade.value_counts()
```

```python
grade_kmeans = grade.copy()
grade_kmeans = grade_kmeans.to_frame(name = 'Kmeans')
grade_kmeans['pred'] = y_pred
grade_kmeans['pred'].value_counts()
```

```python
rep={'A':'2','B':'0','C':'1'}
Y_true_list=[rep[i] if i in rep else i for i in Y_true_list]
```

```python
for i in range(9259):
    Y_true_list[i]
```

```python
sum = 0
for i in range(9259):
    if int(Y_true_list[i]) == y_pred[i]:
        sum = sum + 1
print(sum/9259)
```

```python
grade_kmeans['Kmeans'] = grade_kmeans['Kmeans'].replace({'A':2,'B':0,'C':1})
```

```python
len(grade_kmeans.loc[grade_kmeans['Kmeans'] == grade_kmeans['pred']])/len(grade_kmeans)
```

```python
from sklearn.metrics import precision_score, recall_score

precision_score(grade_kmeans['Kmeans'],grade_kmeans['pred'], average='macro')
```

```python
recall_score(grade_kmeans['Kmeans'],grade_kmeans['pred'], average='macro')
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(grade_kmeans['Kmeans'],grade_kmeans['pred'])
```

```python
print(classification_report(grade_kmeans['Kmeans'],grade_kmeans['pred']))
```
